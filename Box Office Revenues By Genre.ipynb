```python
# Import dependencies
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
```

```python
# Import csv (includes MovieData.csv and API data)
combined_df = pd.read_csv("Resources/api_merge_df.csv")
combined_df
```

```python
# Grouping df by genre
grouped_genre = combined_df.groupby(["genre"])

# Summing totals 
sum_genre = grouped_genre.sum()
sum_genre
```

```python
sum_genre['genre']=sum_genre.index
sum_genre
```

```python
# Bar chart analysis comparing domestic and international box office per genre
x_axis = np.arange(len(sum_genre))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(30,10))
plt.bar(x_axis, sum_genre["domestic_box_office"], color='b', alpha=1, align="center")
plt.xticks(tick_locations, sum_genre["genre"], rotation="horizontal")
plt.title('Domestic box office revenue per genre')
plt.ylabel('Revenue in billions of U.S. dollars')
plt.xlabel('Genres')
```

```python
# Bar chart analysis comparing domestic and international box office per genre
x_axis = np.arange(len(sum_genre))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(30,10))
plt.bar(x_axis, sum_genre["international_box_office"], color='r', alpha=1, align="center")
plt.xticks(tick_locations, sum_genre["genre"], rotation="horizontal")
plt.title('International box office revenue per genre')
plt.ylabel('Revenue in billions of U.S. dollars')
plt.xlabel('Genres')
```

```python
# Bar chart showing total box office revenue per genre
x_axis = np.arange(len(sum_genre))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(30,10))
plt.bar(x_axis, sum_genre["Total Box Office"], color='g', alpha=1, align="center")
plt.xticks(tick_locations, sum_genre["genre"], rotation="horizontal")
plt.title('Total box office revenue per genre')
plt.ylabel('Revenue in billions of U.S. dollars')
plt.xlabel('Genres')
```
