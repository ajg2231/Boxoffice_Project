```python
# Import dependencies
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
```

```python
# Import csv (includes MovieData.csv and API data)
combined_df = pd.read_csv("Resources/api_merge_df.csv")
combined_df
```

```python
# Grouping df by genre
grouped_genre = combined_df.groupby(["genre"])

# Summing totals 
sum_genre = grouped_genre.sum()
sum_genre
```

```python
sum_genre['genre']=sum_genre.index
sum_genre
```

```python
reduced_genre = sum_genre.loc [["Action", "Adventure", "Comedy",
                                "Drama", "Horror", "Romantic Comedy", "Thriller/Suspense"], ["domestic_box_office", "international_box_office"]]
print(reduced_genre)

print()
```

```python
reduced_df = pd.DataFrame(reduced_genre)
reduced_df
```

```python
reduced_df['genre']=reduced_df.index
reduced_df
```

```python
# Bar chart analysis comparing domestic and international box office per genre
x_axis = np.arange(len(reduced_df))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(30,10))
plt.bar(x_axis, reduced_df["domestic_box_office"], color='b', alpha=1, align="center"),  
plt.bar(x_axis, reduced_df["international_box_office"], color='r', alpha=1, align="center")
plt.xticks(tick_locations, reduced_df["genre"], rotation="horizontal")
plt.title('Domestic box office revenue per genre')
plt.ylabel('Revenue in billions of U.S. dollars')
plt.xlabel('Genres')
```
